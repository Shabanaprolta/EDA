# Title: Bike Details Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

#reading the data from csv file
df = pd.read_csv("BIKE DETAILS.csv")

df.head()
df.describe()
```

**Questions:**

```python
#Q1 What is the range of Selling prices in the dataset?

min_price = df['selling_price'].min()
max_price = df['selling_price'].max()
price_range = (min_price , max_price)
print(f"Range of Selling prices : {price_range}")
```

```python
#Q2 What is the median selling price of bikes in the dataset?

median_price = df['selling_price'].median()
print(f"Median Selling Price : {median_price}")
```

```python
#Q3 What is the most common seller type?

common_seller = df['seller_type'].mode()[0]
print(f"Most Common Seller Type : {common_seller}")
```

```python
#Q4 How many bikes have driven more than 50,000 kilometers?

df['km_driven'] = pd.to_numeric(df['km_driven'], errors='coerce')
high_km_bikes = df[df['km_driven'] > 50000].shape[0]
print(f"Number of bikes driven more than 50,000 km : {high_km_bikes}")

```

```python
#Q5 What is the average km_driven valuefor each ownership type?

avg_km_by_owner = df.groupby('owner')['km_driven'].mean()
print(avg_km_by_owner)
```

```python
#Q6  What proportion of bikes are from the year 2015 or older ?

df['year'] = pd.to_numeric(df['year'], errors='coerce')
older_bikes = df[df['year']<=2015].shape[0]
total_bikes = df.shape[0]
proportion = (older_bikes/total_bikes)
print(f"Proportion of bikes from 2015 or older : {proportion:.2%}")
```

```python
#Q7 What is the trend of missing values across the dataset?

missing_values = df.isnull().sum()
print(missing_values)
```

```python
#Q8 What is the highest ex_showroom_price recorded ,and for which bike?

max_ex_price = df['ex_showroom_price'].max()
bike_with_max_ex_price = df[df['ex_showroom_price'] == max_ex_price]['name'].values[0]
print(f"Highest ex_showroom_price : {max_ex_price}, Bike: {bike_with_max_ex_price}")
```

```python
#Q9 What is the total number of bikes listed by each seller type?
bikes_by_seller = df['seller_type'].value_counts()
print(bikes_by_seller)
```

```python
#Q10 What is the relationship between selling_price and km_driven for first - owner bikes?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['selling_price'] = pd.to_numeric(df['selling_price'], errors='coerce')
df['km_driven'] = pd.to_numeric(df['km_driven'], errors='coerce')
if 'owner' in df.columns:
  first_owner_bikes = df[df['owner'].str.lower().str.strip() == '1st owner']
  if not first_owner_bikes.empty:
    plt.figure(figsize=(8, 5))
    sns.regplot(x = first_owner_bikes['km_driven'],y = first_owner_bikes['selling_price'],scatter_kws = {'alpha': 0.5})
    plt.xlabel("Kilometers Driven")
    plt.ylabel("Selling Price(INR)")
    plt.title("Selling Price vs Kilometers Driven For First-Owner Bikes")
    plt.show()
    correlation = first_owner_bikes['selling_price'].corr (first_owner_bikes['km_driven'])
    print(f"Correlation between selling_price and km_driven for first Owner bikes: {correlation: .2f}")
```

```python
#Q11 Identify and remove outliers in the km_driven column using the IQR method.

Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

df_cleared = df[(df['km_driven'] >= Q1 - 1.5 * IQR) & (df['km_driven'] <= Q3 + 1.5 * IQR)]
print(df_cleared)
```

```python
#Q12 Perform a bivariate analysis to visualize the relationship between year and selling_price.

import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='year', y='selling_price')
plt.xticks(rotation=45)
plt.title("Year vs Selling Price")
plt.show()
```

```python
#Q13 What is the average depreciation in selling price based in the bike's age (current year - manufacture year)?

current_year = 2025
df['bike_age'] = current_year - df['year']
df['depreciation'] = df['ex_showroom_price'] - df['selling_price'] /df['bike_age']
avg_depreciation = df['depreciation'].mean()
print(f"Average annual depreciation : {avg_depreciation}")
```

```python
#Q14 Which bike names are priced significantly above the average price for their manufacturing year?

mean_price_by_year = df.groupby('year')['selling_price'].mean()
df_above_avg = df[df['selling_price'] > df['year'].map (mean_price_by_year)]
print(df_above_avg[['name','selling_price']])
```

```python
#Q15 Develop a correlation matrix for numeric columns and visualize it using a heatmap.

import numpy as np

corr_matrix = df[['selling_price','year', 'km_driven', 'ex_showroom_price']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

# **Car Sale Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# reading file from csv
df = pd.read_csv("Car Sale.csv")
df.info()
df.head()
```

```python
# Q1 What is the average selling price of cars for each dealer, and how does it compare across different dealers?

# Calculate the average selling price per dealer
avg_price_per_dealer = df.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending= False)

avg_price_per_dealer.head(10)
```

```python
# Q2 Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends?
price_variation = df.groupby('Company')['Price ($)'].std().sort_values(ascending= False)
price_variation.head(10)
```

```python
#Q3 What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?
df.groupby('Transmission')['Price ($)'].describe()
```

```python
#Q4 What is the distribution of car prices across different regions?
df.groupby("Dealer_Region")["Price ($)"].describe()
```

```python
#Q5 What is the distribution of cars based on body styles?
body_style_counts = df["Body Style"].value_counts()
print(body_style_counts)
```

```python
#Q6 How does the average selling price of cars vary by customer gender and annual income ?
df.groupby('Gender')['Price ($)'].mean()
df.groupby('Annual Income')['Price ($)'].mean()
```

```python
#Q7 What is the distribution of car prices by region , and how does the number of cars sold vary by region?
df.groupby('Dealer_Region')['Price ($)'].mean()
df['Dealer_Region'].value_counts()
```

```python
#Q8 Hoe does the average car price differ between cars with different engine sizes?
df.groupby('Engine')['Price ($)'].mean()
```

```python
#Q9 How do car prices vary based on the customer's annual income bracket?
df.groupby(pd.qcut(df['Annual Income'],4))['Price ($)'].mean()
```

```python
#Q10 What are the top 5 car models with the highest number of sales, and how does their price distribution look?
top_5_models = df['Model'].value_counts().head(5)
df[df['Model'].isin(top_5_models.index)].groupby('Model')['Price ($)'].describe()
```

```python
#Q11 How does car price vary with engine size across different car colors, and which colors have the highest price variation?

df.groupby(['Engine' , 'Color'])['Price ($)'].mean()
```

```python
#Q12 Is there any seasonal trend in car sales based on the date of sale?
df['Date'] = pd.to_datetime(df['Date'],errors = 'coerce')
df['Month'] = df['Date'].dt.month
df['Month'].value_counts().sort_index()
```

```python
#Q13 How does the car price distribution change when considering different combinations of body style and transmission type?
df.groupby(['Body Style','Transmission'])['Price ($)'].describe()
```

```python
#Q14 What is the correlation between car price, engine size, and anuual income of customers, and how do these feature interact?
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')
correlation = df[['Price ($)', 'Engine','Annual Income']].corr()
print(correlation)
```

```python
#Q15 How does the average car price vary across different car models and engine types?
df.groupby(['Model', 'Engine',]) ['Price ($)'].mean().sort_values(ascending = False)
```

# **Amazon Sales Data**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("amazon.csv")

df.head()
```

```python
#Q1 What is the average rating for each product category.
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
avg_rating_per_category = df.groupby('category')['rating'].mean().astype(float)
print(avg_rating_per_category)
```

```python
#Q2 What are the top rating_count products by category?
top_rating_count_products = df.sort_values(by="rating_count", ascending= False)[["category","product_name","rating_count"]].drop_duplicates()
top_rating_count_products.head(10)
```

```python
#Q3 What is the distribution of discounted prices vs. actual prices?
price_distribution = df[["discounted_price","actual_price"]].describe()
print(price_distribution)
```

```python
#Q4 How does the average discount percentage vary across categories?
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.rstrip('%').astype(float))
avg_discount_per_category = df.groupby('category')['discount_percentage'].mean()
print(avg_discount_per_category)
```

```python
#Q5 What are the most popular product names?
top_popular_products = df.sort_values(by="rating_count",ascending = False)[["product_name","rating_count"]].drop_duplicates()
top_popular_products.head(5)
```

```python
#Q6 What are the most popular product keywords ?
from collections import Counter
words = " ".join(df["product_name"].dropna()).lower().split()
common_words = Counter(words).most_common(10)
print(common_words)
```

```python
#Q7 What are the most popular product reviews?
top_popular_reviews = df.sort_values(by="rating_count",ascending = False)[["review_title","review_content"]].drop_duplicates()
top_popular_reviews.head(5)
```

```python
#Q8 What is the correlation between discounted_price and rating?
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('â‚¹','').str.replace(',',''), errors='coerce')
correlation = df ["discounted_price"].corr(df["rating"])
print(correlation)
```

```python
#Q9 What are the top 5 categories based on the highest ratings?
top_5_categories = avg_rating_per_category.sort_values(ascending = False)
top_5_categories.head(5)
```

```python
#Q10 Identify any potential areas for improvement or optimization based on the data analysis.
#low-rated categories need improvement(e.g., better quality, features).
low_rated_categories = df.groupby('category')['rating'].mean().sort_values()
low_rated_categories.head(5)
```

# **Spotify Data: Popular Hip- hop Artists and Tracks**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("spotify.csv")

df.head()
```

```python
#Q1 Read the dataframe , check null value if present then do the needfull , check duplicate row, if present then do the needful.
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates()
```

```python
#Q2 What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.
sns.histplot(df['Popularity'],bins = 10)
```

```python
#Q3 Is there any relationship between the popularity and the duration of tracks ? Explore this using a scatter plot.
sns.scatterplot(x='Duration (ms)',y='Popularity',data = df)
```

```python
#Q4 Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.
df['Artist'].value_counts().idxmax()
plt.figure(figsize=(12,8))
plt.xticks(rotation =45, fontsize = 5)
sns.countplot (x='Artist',data = df,order = df['Artist'].value_counts().index)
plt.show()
```

```python
#Q5 What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.
df.sort_values(by='Popularity').head(5)[['Artist','Track Name']]
```

```python
#Q6 Among the top 5 most popular artists has the highest populatrity in average ? Calculate and display the average popularity for each artist.
df.groupby('Artist')['Popularity'].mean().nlargest(5)
```

```python
#Q7 For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
top_5_artists = df.groupby('Artist')['Popularity'].mean().nlargest(5).index.tolist()
top_tracks_by_artist = df[df['Artist'].isin(top_5_artists)].groupby('Artist')['Track Name'].apply(list)
print(top_tracks_by_artist)
```

```python
#Q8 Visualize relationships between multiple numerical variables simultaneously using a pair plot.
sns.pairplot(df)
```

```python
#Q9 Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.
plt.figure(figsize=(12,8))
plt.xticks(rotation =45, fontsize = 5)
sns.boxplot(x='Artist',y='Duration (ms)',data = df)
plt.show()
```

```python
#Q10 How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.

plt.figure(figsize=(12,8))
plt.xticks(rotation =45, fontsize = 5)
sns.violinplot(x='Artist',y='Popularity',data = df)
plt.show()
```
